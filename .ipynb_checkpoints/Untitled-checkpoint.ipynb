Terlihat `'content-type': 'text/html; charset=UTF-8'`. Karena content type nya adalah `HTML`, sehingga bisa di-parse.

```python
soup = BS(response.content, "html.parser")
```

```python
title = soup.find('title')
print(title.text)
```

```python
rows = []
table = soup.find("table", {'aria-label': 'Core Data'})
table_rows = table.find_all('tr')

for table_row in table_rows:
    td = table_row.find_all('td')
    
    temp_row = []
    for data in td:
        temp_row.append(data.get_text(strip=True))
        
    rows.append(temp_row)
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

gdf = data_gdf.copy()

gdf_sorted = gdf.sort_values(by='indeks_pendidikan', ascending=True)

list_of_gdf_sorted = []

# Iterasi melalui setiap baris di GeoDataFrame
for index, row in gdf_sorted.iterrows():
    # Buat GeoDataFrame baru dari satu baris
    new_gdf_sorted = gpd.GeoDataFrame([row], geometry='geometry', crs=gdf_sorted.crs)
    
    # Tambahkan GeoDataFrame baru ke list
    list_of_gdf_sorted.append(new_gdf_sorted)

# Plot peta Anda dengan warna yang sesuai
fig, ax = plt.subplots(figsize=(24, 12))
gdf.plot(ax=ax, color='lightgray')

legend_labels = []

for i, gdf in enumerate(list_of_gdf_sorted):
    gdf.plot(ax=ax, color=colors[i])
    
    legend_labels.insert(0, plt.Line2D([0], [0], color=colors[i], lw=2, label=gdf['PROVINSI'].values[0]))

ax.legend(handles=legend_labels, title='Provinsi', loc='lower left', bbox_to_anchor=(1.05, 0), borderaxespad=0.)

plt.title('Data Indeks Pendidikan Per Provinsi Se-Indonesia')
plt.axis('off')

plt.show()
```

## 2. Data Rasio Guru dan Siswa SMA Sederajat Berdasarkan Provinsi di Indonesia

### Import Data

```python
def get_table_data(url):
    response = requests.get(url)
    soup = BS(response.content, "html.parser")

    title = soup.find('h3')
    rows = []
    header = []
    table = soup.find("table", {'class': 'table table-styling'})
    table_rows = table.find_all('tr')

    for table_row in table_rows:
        th = table_row.find_all('th')
        td = table_row.find_all('td')

        temp_row = []
        for data in td:
            temp_row.append(data.get_text(strip=True))

        rows.append(temp_row)
        
        header = ['provinsi', 'jumlah sekolah (negeri)', 'jumlah sekolah (swasta)', 'jumlah sekolah (negeri+swasta)',
          'jumlah guru (negeri)', 'jumlah guru (swasta)', 'jumlah guru (negeri+swasta)',
          'jumlah murid (negeri)', 'jumlah murid (swasta)', 'jumlah murid (negeri+swasta)']

    df_rasio = pd.DataFrame(rows[1:35], columns=header)
    
    selected_header = ['provinsi', 'jumlah guru (negeri+swasta)', 'jumlah murid (negeri+swasta)']
    df_rasio_selected = df_rasio[selected_header]
    
    for head in selected_header[1:]:
        df_rasio_selected[head] = df_rasio_selected[head].str.replace('-', '0')
        df_rasio_selected[head] = df_rasio_selected[head].str.replace(' ', '').astype(int)
    
    return response.status_code, title, df_rasio_selected
```

```python
response_sma, title_sma, df_rasio_sma = get_table_data("https://www.bps.go.id/indikator/indikator/view_data_pub/0000/api_pub/a1lFcnlHNXNYMFlueG8xL0ZOZnU0Zz09/da_04/6")
response_smk, title_smk, df_rasio_smk = get_table_data("https://www.bps.go.id/indikator/indikator/view_data_pub/0000/api_pub/MU90V01YZ0RxenhmbFdsU21iUHh2Zz09/da_04/5")
response_ma, title_ma, df_rasio_ma = get_table_data("https://www.bps.go.id/indikator/indikator/view_data_pub/0000/api_pub/UE1YYmk0TnR5TnoycHlYbzNnbkFIZz09/da_04/6")

print(str(response_sma).strip(), ", ", title_sma.text.strip())
print(str(response_smk).strip(), ", ", title_smk.text.strip())
print(str(response_ma).strip(), ", ", title_ma.text.strip())
```

```python
def cleansing_check(df):
    count_nan = df.isnull().sum().sum()
    
    count_zero = (df == 0).sum().sum()

    print(f"Jumlah nilai NaN dalam DataFrame: {count_nan}")
    print(f"Jumlah elemen yang sama dengan 0 dalam DataFrame: {count_zero}")
```

```python
print("SMA")
cleansing_check(df_rasio_sma)

print("\nSMK")
cleansing_check(df_rasio_smk)

print("\nMA")
cleansing_check(df_rasio_ma)
```

```python
# Mencari baris yang memiliki nilai "0" dalam baris tersebut
print(df_rasio_ma[(df_rasio_ma == 0).any(axis=1)])
```

Merujuk pada data tingkat MA, karena terjadi data yang tidak normal pada provinsi yang berpotensi mempengaruhi hasil akhir (yaitu <b>DKI Jakarta</b> dan <b>DI Yogyakarta</b>) maka kami tidak akan mengambil dataframe data tingkat MA. Hal ini tidak cukup mempengaruhi hasil akhir secara signifikan karena hasil akhir adalah rasio antara jumlah guru dan jumlah murid (rasio tidak memperhatikan banyak nya tapi selisihnya).

```python
combined_df = pd.concat([df_rasio_sma, df_rasio_smk])

# Kelompokkan DataFrame gabungan berdasarkan kolom 'provinsi' dan jumlahkan jumlah guru dan jumlah siswa
data_rasio = combined_df.groupby('provinsi').agg({'jumlah guru (negeri+swasta)': 'sum', 'jumlah murid (negeri+swasta)': 'sum'}).reset_index()

data_rasio.head()
```

```python
data_rasio['rasio'] = data_rasio['jumlah guru (negeri+swasta)'] / data_rasio['jumlah murid (negeri+swasta)']

data_rasio = data_rasio[['provinsi', 'rasio']].sort_values(by='provinsi')

data_rasio = data_rasio.reset_index(drop=True)
```

```python
data_rasio.head()
```

```python
data_rasio_sorted = data_rasio.sort_values(by='rasio', ascending=True)
data_rasio_sorted = data_rasio_sorted.reset_index(drop=True)

data_rasio_sorted.head()
```

#### Visualisasi Data

```python
top = data_rasio_sorted.head(17)
bottom = data_rasio_sorted.tail(17)

top_colors = ['lightcoral'] * 17
bottom_colors = ['cornflowerblue'] * 17

combined_data = pd.concat([top, bottom])

combined_colors = top_colors + bottom_colors

provinsi = combined_data['provinsi']
indeks_pendidikan = combined_data['rasio']

plt.figure(figsize=(14, 8))
plt.barh(provinsi, indeks_pendidikan, color=combined_colors)
plt.xlabel('Rasio Guru dan Siswa')
plt.title('Ranking Rasio Guru dan Siswa SMA Sederajat per Provinsi')

plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.show()
```

## 3. Visual Gabungan Data Antara Indeks Pendidikan dan Rasio

```python
# df baru
data_akhir = data_gdf[{'rasio', 'indeks_pendidikan'}].copy()
data_akhir = data_akhir.set_index(data_gdf['PROVINSI'])

# scaling angka

from sklearn.preprocessing import MinMaxScaler

# Melakukan scaling menggunakan MinMaxScaler
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(data_akhir)

# Mengubah kembali hasil scaling menjadi DataFrame
df_scaled = pd.DataFrame(df_scaled, columns=data_akhir.columns, index=data_akhir.index)

print(df_scaled)
```

## Analisis Korelasi

Analisis Menggunakan Analisis Linear Regresi.

```python
# Ambil kolom provinsi, rasio, dan indeks_pendidikan dari data_gdf
provinsi = df_scaled.index
rasio = df_scaled['rasio']
indeks_pendidikan = df_scaled['indeks_pendidikan']

# Atur lebar bar dan koordinat x
lebar_bar = 0.35
x = np.arange(len(provinsi))

# Buat plot
fig, ax = plt.subplots(figsize=(12, 6))
bar1 = ax.bar(x - lebar_bar/2, rasio, lebar_bar, label='Rasio', color='cornflowerblue')
bar2 = ax.bar(x + lebar_bar/2, indeks_pendidikan, lebar_bar, label='Indeks Pendidikan', color='lightcoral')

# Atur label sumbu x
ax.set_xlabel('Provinsi')
ax.set_xticks(x)
ax.set_xticklabels(provinsi, rotation=90)

# Atur label sumbu y
ax.set_ylabel('Nilai')

# Tambahkan judul
ax.set_title('Rasio dan Indeks Pendidikan per Provinsi')

# Tampilkan legenda
ax.legend()

plt.show()
```

```python
df_scaled_sorted = df_scaled.sort_values(by='indeks_pendidikan', ascending=True)

# Ambil kolom provinsi, rasio, dan indeks_pendidikan dari data_gdf
provinsi = df_scaled_sorted.index
rasio = df_scaled_sorted['rasio']
indeks_pendidikan = df_scaled_sorted['indeks_pendidikan']

# Atur lebar bar dan koordinat x
lebar_bar = 0.35
x = np.arange(len(provinsi))

# Buat plot
fig, ax = plt.subplots(figsize=(12, 6))
bar1 = ax.bar(x - lebar_bar/2, rasio, lebar_bar, label='Rasio', color='cornflowerblue')
bar2 = ax.bar(x + lebar_bar/2, indeks_pendidikan, lebar_bar, label='Indeks Pendidikan', color='lightcoral')

# Atur label sumbu x
ax.set_xlabel('Provinsi')
ax.set_xticks(x)
ax.set_xticklabels(provinsi, rotation=90)

# Atur label sumbu y
ax.set_ylabel('Nilai')

# Tambahkan judul
ax.set_title('Rasio dan Indeks Pendidikan per Provinsi')

# Tampilkan legenda
ax.legend()

plt.show()
```
